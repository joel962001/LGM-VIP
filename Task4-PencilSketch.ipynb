```python
# Impoting the required libraries
import cv2
import matplotlib.pyplot as plt
```

```python
# Reading the image
image = cv2.imread("scenery.jpg")
```

```python
# Displaying the orignal image
plt.figure(num = 1, figsize=(8,8))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image', fontsize=26)
plt.gca().axes.get_yaxis().set_visible(False)
plt.gca().axes.get_xaxis().set_visible(False)
plt.show()
```

```python
# Creating a new image by converting the original image to greyscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display image
plt.figure(num = 1, figsize=(8,8))
plt.imshow(gray_image, cmap="gray")
plt.title('Gray Image', fontsize=26)
plt.gca().axes.get_yaxis().set_visible(False)
plt.gca().axes.get_xaxis().set_visible(False)
plt.show()
```

```python
# Inverting the new grayscale image
inverted_image = 255 - gray_image
cv2.imshow("Inverted", inverted_image)

# Display image
plt.figure(num = 1, figsize=(8,8))
plt.imshow(inverted_image, cmap="gray")
plt.title('Inverted Image', fontsize=26)
plt.gca().axes.get_yaxis().set_visible(False)
plt.gca().axes.get_xaxis().set_visible(False)
plt.show()
```

```python
# Blur the image by using the Gaussian Function in OpenCV
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)

# Display image
plt.figure(num = 1, figsize=(8,8))
plt.imshow(blurred, cmap="gray")
plt.title('Blurred Image', fontsize=26)
plt.gca().axes.get_yaxis().set_visible(False)
plt.gca().axes.get_xaxis().set_visible(False)
plt.show()
```

```python
# Invert the blurred image  
inverted_blurred = 255 - blurred

# converting the image into a pencil sketch
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
```

```python
# Display the pencil sketch
plt.figure(num = 1, figsize=(8,8))
plt.imshow(pencil_sketch, cmap="gray")
plt.title('Pencil Sketch', fontsize=26)
plt.gca().axes.get_yaxis().set_visible(False)
plt.gca().axes.get_xaxis().set_visible(False)
plt.show()
```
